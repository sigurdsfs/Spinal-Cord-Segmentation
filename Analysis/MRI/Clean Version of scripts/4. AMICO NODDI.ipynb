# Load Lib

```python
import amico
amico.setup()
```

## Rotated + Transposed

```python
ae_full = amico.Evaluation()
ae_wm_gm = amico.Evaluation()
```

```python
full_SC_mask = "Combined_Data/Rotated and Transposed/DTI_MASK_combined_preproc_trans_masked_rotated.nii.gz"
GM_WM_mask = "Combined_Data/Rotated and Transposed/Combined_GM_WM_mask_aug_rotated_FINAL_f1_threshold_dilated_GM_96x96.nii.gz"
```

```python
ae_full.load_data("Combined_Data/Rotated and Transposed/DTI_combined_preproc_trans_masked_rotated.nii.gz", 
             scheme_filename= "AMICO/scheme.scheme", 
             mask_filename = full_SC_mask, 
             b0_thr=0)
```

Compute the response functions:

Set the NODDI model with the set_model() method and generate the response functions with the generate_kernels() method:

```python
ae_full.set_model('NODDI')
ae_full.generate_kernels(regenerate=True)
ae_full.load_kernels()
ae_full.fit()
ae_full.save_results(path_suffix= "full_mask")
```

```python
ae_wm_gm = amico.Evaluation()

ae_wm_gm.load_data("Combined_Data/Rotated and Transposed/DTI_combined_preproc_trans_masked_rotated.nii.gz", 
             scheme_filename= "AMICO/scheme.scheme", 
             mask_filename = GM_WM_mask, 
             b0_thr=0)

ae_wm_gm.set_model('NODDI')
ae_wm_gm.generate_kernels(regenerate=True)
ae_wm_gm.load_kernels()
ae_wm_gm.fit()
ae_wm_gm.save_results(path_suffix= "wm_gm_mask")
```

```python
#full_SC_mask = "Combined_Data/Transposed/DTI_MASK_combined_preproc_trans_masked_rotated.nii.gz"
GM_WM_trans_mask = "Combined_Data/Transposed/Combined_GM_WM_mask_aug_FINAL_f1_threshold_dilated_96x96.nii.gz"
```
